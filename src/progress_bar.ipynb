```python
from time import sleep
from tqdm import tqdm

for i in tqdm(range(10)):
    sleep(0.5)
```

```python
from time import sleep
from tqdm import trange

for i in trange(10):
    sleep(0.5)
```

```python
import time  
from tqdm import tqdm

names = ["Alice", "Bob", "Charlie", "David", "Eve"]

pbar = tqdm(names)
for name in pbar:
    pbar.set_description(f"Processing {name}")
    print(pbar.n)
    time.sleep(1)
```
